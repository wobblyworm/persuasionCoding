```python
def getCode_from_txt(txt):
    match txt:
          case "Rational Persuasion":
              return 10
          case "Logical Appeal":
              return 11
          case "Empirical Evidence":
              return 12
          case "Personal Disclosure":
              return 13
          case "Persuasive Inquiry":
              return 14
          case "Manipulation":
              return 20
          case "Social Conformity":
              return 21
          case "Negative Emotional Appeal":
              return 22
          case "Scarcity":
              return 23
          case "Status Quo Bias":
              return 24
          case "Othering":
              return 25
          case "Authority Appeal":
              return 26
          case "Others (M)":
              return 27
          case "Others (RP)":
              return 15
          case "Not Persuasion":
              return 0
          case "N/A":
              return 1
          case _:
              return -1

def getCode_from_AbbrevTxt(txt):
  match txt:
        case "RP":
            return 10
        case "LA":
            return 11
        case "EE":
            return 12
        case "PD":
            return 13
        case "PI":
            return 14
        case "M":
            return 20
        case "SC":
            return 21
        case "NEA":
            return 22
        case "S":
            return 23
        case "SQB":
            return 24
        case "O":
            return 25
        case "AA":
            return 26
        case "O(M)":
            return 27
        case "O(RP)":
            return 15
        case "NP":
            return 0
        case "N/A":
            return 1
        case _:
            return -1

def getTxt_from_AbbrevTxt(code):
  match code:
        case "NP":
              return "Not Persuasion"
        case "N/A":
                return "N/A"
        case "O(M)":
                return "Others (M)"
        case "O(RP)":
                  return "Others (RP)"
        case "RP":
              return "Rational Persuasion"
        case "LA":
              return "Logical Appeal"
        case "EE":
              return "Empirical Evidence"
        case "PD":
              return "Personal Disclosure"
        case "PI":
              return "Persuasive Inquiry"
        case "M":
              return "Manipulation"
        case "SC":
              return "Social Conformity"
        case "NEA":
              return "Negative Emotional Appeal"
        case "S":
              return "Scarcity"
        case "SQB":
              return "Status Quo Bias"
        case "O":
              return "Othering"
        case "AA":
              return "Authority Appeal"
        case _:
              return "??"

def getTxt_from_code(code):
  match int(code):
    case 0:
      return "Not Persuasion"
    case 1:
        return "N/A"
    case 27:
        return "Others (M)"
    case 15:
          return "Others (RP)"
    case 10:
      return "Rational Persuasion"
    case 11:
      return "Logical Appeal"
    case 12:
      return "Empirical Evidence"
    case 13:
      return "Personal Disclosure"
    case 14:
      return "Persuasive Inquiry"
    case 20:
      return "Manipulation"
    case 21:
      return "Social Conformity"
    case 22:
      return "Negative Emotional Appeal"
    case 23:
      return "Scarcity"
    case 24:
      return "Status Quo Bias"
    case 25:
      return "Othering"
    case 26:
      return "Authority Appeal"
    case _:
      return "??"

def getTxtAbbrev_from_txt(txt):
  match txt:
        case "Rational Persuasion":
            return "RP"
        case "Logical Appeal":
            return "LA"
        case "Empirical Evidence":
            return "EE"
        case "Personal Disclosure":
            return "PD"
        case "Persuasive Inquiry":
            return "PI"
        case "Manipulation":
            return "M"
        case "Social Conformity":
            return "SC"
        case "Negative Emotional Appeal":
            return "NEA"
        case "Scarcity":
            return "S"
        case "Status Quo Bias":
            return "SQB"
        case "Othering":
            return "O"
        case "Authority Appeal":
            return "AA"
        case "Others (M)":
            return "OS(M)"
        case "Others (RP)":
            return "OS(RP)"
        case "Not Persuasion":
            return "NP"
        case "N/A":
            return "N/A"
        case _:
            return "??"

def getTxtAbbrev_from_code(code):
  if pd.isna(code):
    return np.nan
  else:
    match int(code):
      case 0:
        return "NP"
      case 1:
        return "N/A"
      case 15:
          return "OS(RP)"
      case 27:
          return "OS(M)"
      case 10:
        return "RP"
      case 11:
        return "LA"
      case 12:
        return "EE"
      case 13:
        return "PD"
      case 14:
        return "PI"
      case 20:
        return "M"
      case 21:
        return "SC"
      case 22:
        return "NEA"
      case 23:
        return "S"
      case 24:
        return "SQB"
      case 25:
        return "O"
      case 26:
        return "AA"
      case _:
        return "??"

def getTxtAbbrevHaein_from_code(code):
  if pd.isna(code):
    return np.nan
  else:
    match int(code):
      case 0:
        return "NP"
      case 1:
        return "N/A"
      case 15:
          return "OS"
      case 27:
          return "OS"
      case 10:
        return "RP"
      case 11:
        return "LA"
      case 12:
        return "EE"
      case 13:
        return "PD"
      case 14:
        return "PI"
      case 20:
        return "M"
      case 21:
        return "SC"
      case 22:
        return "NEA"
      case 23:
        return "S"
      case 24:
        return "SQB"
      case 25:
        return "O"
      case 26:
        return "AA"
      case _:
        return "??"
```

```python
# Create a number input widget
options1 = [('missing',-1), ('Not Persuasion (NP)', 0), ('Rational Persuasion (RP)', 10),
            ('Manipulation (M)', 20)]
options2_RP = [('missing',-1), ('Logical Appeal (LA)', 11), ('Empirical Evidence (EE)', 12),
               ('Personal Disclosure (PD)', 13), ('Persuasive Inquiry (PI)', 14),
               ('Others (OS)', 15)]
options2_M = [('missing',-1), ('Social Confirmity (SC)', 21), ('Negative Emotion Appeal (NEA)', 22),
              ('Scarcity (S)', 23), ('Status Quo Bias (SQB)', 24), ('Othering (O)',25),
              ('Authority Appeal (AA)', 26), ('Others (OS)', 27)]
options2 = [('missing',-1),('N/A',1)]

in_code1 = widgets.Dropdown(
    options = options1,
    value = -1,
    description = '1 Lev Code:',
    disabled = False)

in_code2 = widgets.Dropdown(
    options = options2,
    value = -1,
    description = '2 Lev Code:',
    disabled = False,
)


in_note = widgets.Textarea(
    description='Note:',
    layout=widgets.Layout(width='600px', height='100px')
)

# Output area
output = widgets.Output()

# Text widgets for editing
txt = widgets.Textarea(
    description='Text:',
    layout=widgets.Layout(width='600px', height='100px')
)

# in_code1 = widgets.Text(description="1 Lev Code:")
# in_code2 = widgets.Dropdown(description="1 Lev Code:")

# Button to apply updates
update_button = widgets.Button(description="Update Row")
check_button = widgets.Button(description="Check Codes")


# Display the widget
# display(txt, in_code1, in_code2, in_note, update_button, check_button, output)


# Function to populate text inputs with current row values
def populate_fields(change):
    # global in_note
    index = row_dropdown.value
    txt.value = str(dat.loc[index, 'text'])
    if not pd.isna(dat.loc[index,'first']):
      in_code1.value = int(dat.loc[index,'first'])
    else:
      in_code1.value = -1
    if not pd.isna(dat.loc[index,'second']):
      in_code2.value = int(dat.loc[index,'second'])
    else:
      in_code2.value = -1
    if not pd.isna(dat.loc[index,'note']):
      in_note.value = str(dat.loc[index,'note'])
    else:
      in_note.value = ""

# Function to update row
def update_row(b):
    index = row_dropdown.value
    dat.at[index,'first'] = str(in_code1.value)
    dat.at[index,'second'] = str(in_code2.value)
    dat.at[index, 'note'] = str(in_note.value)

    # index = row_dropdown.value
    # if int(in_code2.value) == 0:
    #   dat.at[index,'first'] = ""
    # elif (int(in_code2.value) < 20):
    #   dat.at[index, 'first'] = "10"
    # else:
    #   dat.at[index, 'first'] = "20"
    # dat.at[index, 'note'] = str(in_note.value)

    with output:
        clear_output()
        print("DataFrame updated:")
        display(dat.loc[index])


# Function to check codes
def check_codes(b):
    # global dat
    with output:
        clear_output()
        print("\nlevel 1 check (uncoded, missing): (", dat['first'].isna().sum(),", ", (dat['first'] == "-1").sum(), ")", sep="")
        print("\nlevel 2 check (uncoded, missing): (", dat['second'].isna().sum(),", ", (dat['second'] == "-1").sum(), ")", sep="")
        print("sum of rational persuation: ",(dat['first'] == 10).sum())
        print("sum of manipulation: ",(dat['first'] == 20).sum())

        # display(df)


def update_code1(change):
  # global in_code1
  val = int(in_code1.value)
  if val == 0:
      in_code2.options = options2
      in_code2.value = 1
  elif val == 10:
      in_code2.options = options2_RP
      in_code2.value = -1
  elif val == 20:
      in_code2.options = options2_M
      in_code2.value = -1
  else:
      in_code2.options = options2
      in_code2.value = -1

# Button click event
update_button.on_click(update_row)
check_button.on_click(check_codes)

# Display UI
# Function to format dropdown labels
def row_label(i):
    return f"Row: {i}, ID: {dat.loc[i, 'custom_id']}"

# Dropdown to show current row info
row_dropdown = widgets.Dropdown(
    options=[(row_label(i), i) for i in dat.index],
    value=0,
    description='Current Row:'
)

# Automatically populate fields when row is selected
row_dropdown.observe(populate_fields, names='value')
in_code1.observe(update_code1, names='value')


# Buttons to move up/down
prev_button = widgets.Button(description='⬆️ Up')
next_button = widgets.Button(description='⬇️ Down')

# Function to update dropdown based on current value
def update_dropdown(index):
    row_dropdown.options = [(row_label(i), i) for i in dat.index]
    row_dropdown.value = index

# Button functions
def on_prev_clicked(b):
    if row_dropdown.value > 0:
        update_dropdown(row_dropdown.value - 1)

def on_next_clicked(b):
    if row_dropdown.value < len(dat) - 1:
        update_dropdown(row_dropdown.value + 1)

# Bind buttons
prev_button.on_click(on_prev_clicked)
next_button.on_click(on_next_clicked)

# Layout and display
display(widgets.HBox([prev_button, next_button]), row_dropdown, txt, in_code1, in_code2, in_note, update_button, check_button, output)
```
